```python
import pickle
import pandas as pd
import numpy as np
import random
```

```python
#Load All files

with open('best_model.pkl', 'rb') as file_1:
  model_ada_best = pickle.load(file_1)
```

# Membuat Data Dummy

```python
# Mengatur seed
seed_value = 777
random.seed(seed_value)
np.random.seed(seed_value)

# Jumlah data point yang ingin dibuat
n = 100

age = np.random.uniform(0.08, 82.0, n)
avg_glucose_level = np.random.uniform(55.22, 169.365, n)
bmi = np.random.uniform(10.3, 47.5,n)

hypertension = [0, 1]
hypertension = random.choices(hypertension, k=n)

heart_disease = [0, 1]
heart_disease = random.choices(heart_disease, k=n)

ever_married = ['No', 'Yes']
ever_married = random.choices(ever_married, k=n)


# Membuat DataFrame
df_inference = pd.DataFrame({
    'age': age,
    'avg_glucose_level': avg_glucose_level,
    'bmi': bmi,
    'hypertension': hypertension,
    'heart_disease': heart_disease,
    'ever_married': ever_married   
})
```

```python
df_inference
```

```python
inf_predict = model_ada_best.predict(df_inference)
inf_predict
```

```python
final = pd.DataFrame(inf_predict, columns=['predict'])
final
```

```python
# Membuat variabel baru dengan menggabungkan 2 tabel yaitu df_inference dan final
tabel_inf = pd.concat([df_inference, final], axis=1)
tabel_inf.head(20)
```
